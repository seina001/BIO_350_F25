```python
#Create an algorithm that calculates the population of 3 ages of a beetle population and the total population over 50 years
# The initial First year population is 500 and the initial second year population is 500
# On average 1 beetle produces 1 offspring in second year and 5 offspring in third year
# The first year beetles survive 30% to second year and 50% of second year beetles survive to third year

#Create a definition to calculate and return the population of each age group and the total population for each year with an initial population of 500 first year beetles and 500 second year beetles
def beetle_population_withtrap(years, initial_first_year, initial_second_year):
    """
    Calculate the population of each age group and the total population for each year.

    Parameters
    ------
    years : integer
        Number of years to calculate the population for.
    initial_first_year : integer
        Initial population of first-year beetles.
    initial_second_year : integer
        Initial population of second-year beetles.

        
    Returns
    -------
    first_year_population : list
        Population of first-year beetles for each year.
    second_year_population : list
        Population of second-year beetles for each year.
    third_year_population : list
        Population of third-year beetles for each year.
    total_population : list
        Total population of beetles for each year.
    """

    #Complete the algorithm using matrix multiplication
    # Create a matrix to represent the transition of populations between years
    # Adjust transition matrix to account for traps catching 5% of beetles each year
    trap_rate = 0.05
    A = np.array([[0, 1, 5],
                  [0.3, 0, 0],
                  [0, 0.5, 0]]) * (1 - trap_rate)
    # Initialize the population vector with the initial populations
    initial_population_vector = np.array([initial_first_year, initial_second_year, 0])
    # Create lists to store the populations of each age group and total population for each year
    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []
    # Append the current populations to the lists
    first_year_population.append(initial_population_vector[0])
    second_year_population.append(initial_population_vector[1])
    third_year_population.append(initial_population_vector[2])
    total_population.append(sum(initial_population_vector))
    # Loop through the number of years to calculate the populations
    n_t = initial_population_vector
    for year in range(years):
        # Calculate the population for the next year using matrix multiplication
        n_t = np.dot(A, n_t)
        # Append the current populations to the lists
        first_year_population.append(n_t[0])
        second_year_population.append(n_t[1])
        third_year_population.append(n_t[2])
        total_population.append(sum(n_t))
    # Return the populations of each age group and total population for each year
    return first_year_population, second_year_population, third_year_population, total_population

```

```python
first_year_pop, second_year_pop, third_year_pop, total_pop = beetle_population_withtrap(150, 500, 500)
```

```python
#Plot the populations of each age group and total population over the years using matplotlib

plt.figure(figsize=(12, 8))
plt.plot(first_year_pop, label='First Year Beetles', color='blue')
plt.plot(second_year_pop, label='Second Year Beetles', color='orange')
plt.plot(third_year_pop, label='Third Year Beetles', color='green')
plt.plot(total_pop, label='Total Population', color='red', linestyle='--')
plt.title('Beetle Population Over Time')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()
```
