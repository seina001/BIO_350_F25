```python
# Import numpy 
import numpy as np
import matplotlib.pyplot as plt
```

1. What is the stationary state distribution of this population?
2. What is the proability that a randomely selected individual from this population at year 50 is a Maple?


```python
# Create function to calculate tree population over time using matrix multiplication
def tree_population(initial_population, years):
    #Initialize the transition Matrix A
    A = np.array([[0.8, 0.25, 0.1], [0.1, 0.15, 0.25], [0.1, 0.6, 0.65]])
    # Initialize the population vector with the initial populations
    n_t = initial_population
    print(n_t)
    # Create lists to store the populations of each age group and total population for each year
    BeechTrees = []
    MapleTrees = []
    Others = []
    # Append the current populations to the lists
    BeechTrees.append(n_t[0])
    MapleTrees.append(n_t[1])
    Others.append(n_t[2])
    # Loop through the number of years to calculate the populations
    for _ in range(years):
        # Calculate the population for the next year using matrix multiplication
        n_t = np.dot(A, n_t)
        print(n_t)
        # Append the current populations to the lists
        BeechTrees.append(n_t[0])
        MapleTrees.append(n_t[1])
        Others.append(n_t[2])
    # Return the populations of each age group and total population for each year
    return BeechTrees, MapleTrees, Others
```

```python
# Plot populations produced by tree_population()

years = 50
initial_population = np.array([1, 20, 79])

BeechTrees, MapleTrees, Others = tree_population(initial_population, years)

print(MapleTrees)

```

```python

plt.figure(figsize=(12, 8))
plt.plot(BeechTrees, label='Beech Trees', color='brown', marker='o')
plt.plot(MapleTrees, label='Maple Trees', color='green')
plt.plot(Others, label='Other Trees', color='gray')
plt.title('Tree Population Over Time')
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
