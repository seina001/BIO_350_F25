```python
#Import matplotlib library
import matplotlib.pyplot as plt
```

```python
def N_exp(N_0, p, time):
    # Exponential growth model
    # Save N_values into a list
    N_values = [N_0]
    # Save initial population
    current_N = N_0
    for _ in range(time):
        current_N = current_N + p * current_N
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 40 days and p = 0.4 with an initial population of 0.01
result = N_exp(0.01, 0.4, 35)
print(result)
```

```python
# Plot the results with days on the x-axis and population on the y-axis
log_result = log(result)
days = list(range(36))  # 0 to 35 days
plt.plot(days, log_result), marker='o')
plt.xlabel('Days')
plt.ylabel('Duckweed Population')
plt.title('Duckweed Population Growth Over Time')
plt.grid()
plt.show()
```

```python
#Change N_Exp to N_exp2 and have 3 conditions based on N_0
#If N_0 < 500, p = 0.4
#If N_0 >= 500 and N_0 < 1000, p = 0.2
#If N_0 >= 1000, p = 0

def N_exp2(N_0, time):
    # Exponential growth model with variable p based on N_0
    # Save N_values into a list
    N_values = [N_0]
    # Save initial population
    current_N = N_0
    for _ in range(time):
        if current_N < 500:
            p = 0.4
        elif 500 <= current_N < 1000:
            p = 0.2
        else:
            p = 0
        current_N = current_N + p * current_N
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 35 days with an initial population of 0.01
result = N_exp2(1, 35)
print(result)
```

```python
# Plot the results with days on the x-axis and population on the y-axis

days = list(range(36))  # 0 to 35 days
plt.plot(days, result, marker='o')
plt.xlabel('Days')
plt.ylabel('Duckweed Population')
plt.title('Duckweed Population Growth Over Time')
plt.grid()
plt.show()
```
